### Reading JSON
+ JSON
+ Pandas

```python
import json
df_json = pd.read_json('myvalidjsonfile.json')
```

```python
df_json.columns
```

```python
with open('myvalidjsonfile.json','r') as jsd:
    json_data = json.load(jsd)
    print(json_data)
```

```python
from pprint import pprint as pp
pp(json_data)
```

#### CONVERT JSON TO DF

```python
json_data.keys()
```

```python
from pandas.io.json import json_normalize
df_json = json_normalize(json_data['Employee'])
```

```python
df_json
```

####  OTHER JSON SAMPLES

```python
import json
with open('sales.json','r') as jsd:
    json_data = json.load(jsd)
    print(json_data)
```

```python
from pprint import pprint as pp
pp(json_data)
```

```python
df_json = json_normalize(json_data['contents'], 'monthlySales', ['category', 'region'])
df_json.head()
```

#### READ USING BEAUTIFUL SOUP

```python
import urllib.request
import lxml
from bs4 import BeautifulSoup
url = 'http://www.fdic.gov/bank/individual/failed/banklist.html'

page = urllib.request.urlopen(url)
soup = BeautifulSoup(page,'lxml')
```

```python
soup.table
```
