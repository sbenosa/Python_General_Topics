## MERGE

```python
import pandas as pd
weather = pd.DataFrame({'City': ['Manila','Cebu','Davao','Solano','Baguio'],
                'Temperature':[30,25,22,18,12],
                'Humidity':[20,25,15,11,10]    
})
weather
```

```python
import pandas as pd
rainfall = pd.DataFrame({'City': ['Manila','Cebu','Davao','Solano','Baguio','Bohol'],
                'Windspeed':[11,10,14,29,22,45],
                'Rainfall':[10,125,115,111,140,29]    
})
rainfall
```

Only the columns that exist on both dataframes will be merged; the default is 'inner'. The indicator will display another field to show the origing of values

```python
pd.merge(weather,rainfall,on='City',how='inner')
```

```python
pd.merge(weather,rainfall,on='City',how='outer',indicator=True)
```

## CONCAT

```python
import pandas as pd
asean = pd.DataFrame({'City': ['Manila','Singapore','Kuala Lumpur','Jakarta','Hanoi'],
                'Temperature':[30,25,22,18,12],
                'Humidity':[20,25,15,11,10]    
})

nap = pd.DataFrame({'City': ['Tokyo','Hong Kong','Seoul'],
                'Temperature':[25,20,18],
                'Humidity':[15,21,18]    
})
```

```python
asean
```

```python
nap
```

```python
pd.concat([asean,nap],ignore_index=True)
```

```python
dataset = pd.concat([asean,nap],keys=["ASEAN","NAP"])
dataset
```

```python
dataset.loc['NAP']
```

#### ADD NEW COLUMN

```python
dataset = pd.concat([nap,nap_ws],axis=1)
dataset
```
